```python
def get_video_ids(youtube, playlist_id):
    
    request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId = playlist_id,
                maxResults = 50)
    response = request.execute()
    
    video_ids = []
    
    for i in range(len(response['items'])):
        video_ids.append(response['items'][i]['contentDetails']['videoId'])
        
    next_page_token = response.get('nextPageToken')
    more_pages = True
    
    while more_pages:
        if next_page_token is None:
            more_pages = False
        else:
            request = youtube.playlistItems().list(
                        part='contentDetails',
                        playlistId = playlist_id,
                        maxResults = 50,
                        pageToken = next_page_token)
            response = request.execute()
    
            for i in range(len(response['items'])):
                video_ids.append(response['items'][i]['contentDetails']['videoId'])
            
            next_page_token = response.get('nextPageToken')
        
    return video_ids
```

```python
def get_video_details(youtube, video_ids):
    all_video_stats = []
    
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
                    part='snippet,statistics',
                    id=','.join(video_ids[i:i+50]))
        response = request.execute()
        
        for video in response['items']:
            video_stats = dict(Title = video['snippet']['title'],
                               Published_date = video['snippet']['publishedAt'],
                               Views = video['statistics']['viewCount'],
                               Likes = video['statistics']['likeCount'],
                               )
            all_video_stats.append(video_stats)
    
    return all_video_stats
```

```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

### Coco-melon topic modeling

```python
# Part 1
# Initializing the lemmatizer
lemma = nltk.stem.WordNetLemmatizer()
```

```python
c1=video_data3['Title'].astype('str')
```

```python
c1
```

```python
step1_output = []
step3_output = []
count = 1
for doc in c1:
    # Tokenizing (Step 1)
    token_doc = nltk.word_tokenize(doc.lower())
    step1_output.append(token_doc)
    print("Step1 done for doc ", count)
    # Lemmatizing and removing anything other than alphabets(Step 2)
    lemma_token_doc = [lemma.lemmatize(token) for token in token_doc if token.isalpha()]
    print("Step2 done for doc ", count)
    # Stop word removal (Step 3)
    stop_words_removed = [token for token in lemma_token_doc if not token in stopwords.words('english') if
                             token.isalpha()]
    step3_output.append(stop_words_removed)
    print("Step3 done for doc ", count)
    count += 1
# Converting the output in step 3 from list of list of strings to list of sentences
step3_output = [' '.join(i) for i in step3_output]
print("Successfully run step 1, step 2 and step 3")
```

```python
step3_output
```

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA
```

```python
# Initializing the tf-idf vectorizers for steps 4 and 5 respectively
vec = TfidfVectorizer(ngram_range=(1, 2), min_df=5)
# Fit and transform the step 3 output
vec_output = vec.fit_transform(step3_output)
lda = LDA(n_components=6).fit(vec_output)

terms = vec.get_feature_names()
for topic_idx, topic in enumerate(lda.components_):
    print("Topic %d:" % (topic_idx))
    print(" ".join([terms[i] for i in topic.argsort()[:-6:-1]]))
```

```python
topic_output = lda.transform(vec_output)
video_data3['Topic'] = topic_output.argmax(axis=1)

print(lda.components_.shape)
print(video_data3[720:])
print(video_data3[:10])
```
